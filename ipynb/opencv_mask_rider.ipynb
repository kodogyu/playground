```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_filename = "/home/kodogyu/Datasets/PAIR360/0924_woojungwon/fisheye/origin1/0000.png"
image = cv2.imread(image_filename, cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(8, 12))
plt.imshow(image, cmap='gray')
plt.show()
```

```python
masked_image = cv2.ellipse(img=image.copy(), center=(1500,3500), axes=(600, 960), angle=60, startAngle=0, endAngle=360, color=0, thickness=-1)

plt.figure(figsize=(8,12))
plt.imshow(masked_image, cmap='gray')
plt.show()
```
