```python
dataset_path = "/home/kodogyu/Datasets/PAIR360/0924_woojungwon/fisheye/origin1/"

images = glob(dataset_path + "*.png")
images.sort()
print(images[:3])
```

```python
image = cv2.imread(images[0])

plt.imshow(image)
plt.show()
```

```python
rotated_output_path = "/home/kodogyu/Datasets/PAIR360/0924_woojungwon/fisheye/rotated1/"

for image_filename in images:
    image_name = image_filename.split('/')[-1]

    image = cv2.imread(image_filename)

    rotated_img = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)

    print(image_name, cv2.imwrite(os.path.join(rotated_output_path, image_name), rotated_img))
```
