```python
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3D 점 데이터 생성
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# 각 점에 대한 색상값 생성 (임의로 생성)
colors = np.random.rand(100)

# 3D scatter plot 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c=colors, cmap='viridis')

# colorbar 추가
plt.colorbar(scatter)

# 플롯 축 레이블 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# 플롯 제목 설정
plt.title('3D Scatter Plot with Colors')

plt.show()
```

```python
# 3D 점 데이터 생성
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# 각 점에 대한 RGB 형태의 색상값 생성 (0부터 1까지의 값)
colors = np.random.rand(100, 3)  # 각 행은 RGB 값

# 3D scatter plot 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c=colors)

# colorbar 추가
plt.colorbar(scatter)

# 플롯 축 레이블 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# 플롯 제목 설정
plt.title('3D Scatter Plot with RGB Colors')

plt.show()
```
