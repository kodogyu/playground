```python
import os
import csv

from copy import deepcopy
import numpy as np
import cv2
import matplotlib.pyplot as plt

import rosbag
from sensor_msgs.msg import Image
from cv_bridge import CvBridge
```

```python
def extract_frames(topic, image, interested_frames, frame_ids):
    bridge = CvBridge()
    # left frame
    if topic == "/camera/color/image_raw":
        if str(image.header.stamp.secs) + str(image.header.stamp.nsecs) in interested_frames:
            left_image = bridge.imgmsg_to_cv2(image, "bgr8")
            cv2.imwrite("../test_image_sequences/l515_aligned_vertical_stereo_kf/left_frames/left_image" + str(image.header.seq) + ".png", left_image)
            print("left_image" + str(image.header.seq) + ".png")
            frame_ids.append(image.header.seq)
    # right frame
    elif topic == "/camera/color_right/image_raw":
        if str(image.header.stamp.secs) + str(image.header.stamp.nsecs) in interested_frames:
            right_image = bridge.imgmsg_to_cv2(image, "bgr8")
            cv2.imwrite("../test_image_sequences/l515_aligned_vertical_stereo_kf/right_frames/right_image" + str(image.header.seq) + ".png", right_image)
            print("right_image" + str(image.header.seq) + ".png")
            # frame_ids.append(image.header.seq)
```

```python
# interesting frames
interested_frames_timestamp = []

def read_csv_file(file_path):
    with open(file_path, 'r', newline='') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            interested_frames_timestamp.append(row['#timestamp_lkf'])

csv_file_path = '/home/kodogyu/kimera_multi_ws/src/kimera_ros/output_logs/L515_stereo/output_frontend_stats.csv'
read_csv_file(csv_file_path)


print(len(interested_frames_timestamp), "frames in total")
```

```python
# Read rosbag
read_bag_filename = "/home/kodogyu/Datasets/rosbags/l515_aligned_vertical_stereo.bag"
read_bag = rosbag.Bag(read_bag_filename)
interested_frames_id = []

topics = ['/camera/color/image_raw', '/camera/color_right/image_raw']
for topic, image, t in read_bag.read_messages(topics=topics):
    extract_frames(topic, image, interested_frames_timestamp, interested_frames_id)
    # print(topic, image.header.seq - 477)

read_bag.close()
```

```python
# write frame ids to csv file
print(interested_frames_id)
with open('files/interesting_frames.csv', 'w') as file:
    for frame_id in interested_frames_id:
        file.write(str(frame_id) + '\n')
```
