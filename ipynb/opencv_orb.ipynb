```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_file = "/home/kodogyu/Pictures/kitti_test_seq_10/000000.png"

image = cv2.imread(image_file, cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(15,10))
plt.imshow(image, cmap='gray')
plt.show()
```

```python
orb = cv2.ORB_create(nfeatures=2000, scaleFactor=1.2, nlevels=8, edgeThreshold=31, firstLevel=0, WTA_K=2, scoreType=cv2.ORB_FAST_SCORE, patchSize=31, fastThreshold=20)

kps = orb.detect(image)

kp_image = cv2.drawKeypoints(image, kps, None)

plt.figure(figsize=(15,10))
plt.imshow(kp_image)
plt.show()
```

```python
cv2.imwrite("/home/kodogyu/Pictures/orb_features_2000.png", kp_image)
```
