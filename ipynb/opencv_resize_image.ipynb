```python
import os
from glob import glob

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_filename = "/home/kodogyu/Datasets/PAIR360/0924_woojungwon/fisheye/rotated1/0000.png"
image = cv2.imread(image_filename)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.show()
```

```python
(h, w) = image.shape[:2]

resized_img = cv2.resize(image, (w//2, h//2))

plt.imshow(resized_img)
plt.show()
```

```python
dataset_path = "/home/kodogyu/Datasets/PAIR360/0924_woojungwon/fisheye/rotated1/"

images = glob(dataset_path + "*.png")
images.sort()
print(images[:3])
```

```python
resized_output_path = "/home/kodogyu/Datasets/PAIR360/0924_woojungwon/fisheye/resized1/"

for image_filename in images[:100]:
    image_name = image_filename.split('/')[-1]

    image = cv2.imread(image_filename)

    resized_img = cv2.resize(image, (w//2, h//2))

    print(image_name, cv2.imwrite(os.path.join(resized_output_path, image_name), resized_img))
```
